```python
# setup paths
non_slip_path = '../dataset3/0'
slip_path = '../dataset3/1'

target_non_slip_path = '../dataset3/processed/0'
target_slip_path = '../dataset3/processed/1'

if not os.path.exists(target_non_slip_path):
    os.makedirs(target_non_slip_path)
if not os.path.exists(target_slip_path):
    os.makedirs(target_slip_path)
```

```python
frame_names_0 = glob.glob('../dataset3/0/*')
frame_names_1 = glob.glob('../dataset3/1/*')
```

```python
# readin csv file and transform them into images for visualization


Nx, Ny = 30, 30

for path in frame_names_0:
    frames = pd.read_csv(path)
    folder_name = os.path.splitext(os.path.basename(path))[0]
    folder_path = os.path.join(target_non_slip_path, folder_name)
    
    # make folder for each set of frames
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
    
    data = frames.values[:,1:]
    print data[:, 1].shape
    feature_size = Nx*Ny
    for i in range(0, data.shape[1]):
        dx_interp = data[:feature_size, i]
        dy_interp = data[feature_size:, i]
#         print dx_interp.shape, dy_interp.shape
        
        mag = np.sqrt(dx_interp ** 2 + dy_interp ** 2)

        dx_resized = (dx_interp.reshape(Nx, Ny) + 30) * 256 / 60
        dy_resized = (dy_interp.reshape(Nx, Ny) + 30) * 256 / 60
        mag_resized = (mag.reshape(Nx, Ny) + 30) * 256 / 60
        # print '====== ', dx_resized.max(), dx_resized.min()

#         # fx, fy = 1, 1
#         dx_large = cv2.resize(dx_resized, (0, 0), fx=fx, fy=fy, interpolation=cv2.INTER_LINEAR)
#         dy_large = cv2.resize(dy_resized, (0, 0), fx=fx, fy=fy, interpolation=cv2.INTER_LINEAR)
#         mag_large = cv2.resize(mag_resized, (0, 0), fx=fx, fy=fy, interpolation=cv2.INTER_LINEAR)

        # concated = np.concatenate((dx_large, dy_large, mag_large), axis=1)
        # concated = np.array(concated, dtype=np.uint8)
        dx_large = np.array(dx_resized, dtype=np.uint8)
        dy_large = np.array(dy_resized, dtype=np.uint8)
        mag_large = np.array(mag_resized, dtype=np.uint8)

        dx_large = cv2.applyColorMap(dx_large, cv2.COLORMAP_JET)
        dy_large = cv2.applyColorMap(dy_large, cv2.COLORMAP_JET)
        mag_large = cv2.applyColorMap(mag_large, cv2.COLORMAP_JET)
        
#         print dx_large
#         print 
        cv2.imwrite(os.path.join(folder_path, '{}_dx.jpg'.format(i)), dx_large)
        cv2.imwrite(os.path.join(folder_path, '{}_dy.jpg'.format(i)), dy_large)
        cv2.imwrite(os.path.join(folder_path, '{}_mag.jpg'.format(i)), mag_large)
        
        
#     break
        
   
```

```python
import cv2

def vec_color_encoding(x, y, encoding='hsv'):
    if not x.shape == y.shape:
        print '2d vector components should have same shapes.'
        return None
    hsv = np.zeros((x.shape[0], x.shape[1], 3))
    hsv[..., 1] = 255

    mag, ang = cv2.cartToPolar(x, y)
    
    # print np.max(mag), np.min(mag)
    hsv[..., 0] = ang*180/np.pi/2
    # hsv[..., 2] = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)
    hsv[..., 2] = mag*10
    
    hsv = np.uint8(hsv)
    if encoding == 'hsv':
        return hsv
    elif encoding == 'rgb':
        # print('converted')
        bgr = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
        return bgr
```

## visualization
we visualize the frames in each data point with optical flow like vector representation.


```python
Nx, Ny = 30, 30

ri = np.random.randint(0, 999)
print ri
path = glob.glob(to_dict['r_path'] + '/*')[625]

frames = pd.read_csv(path)


data = frames.values
print data[:, 1].shape
feature_size = Nx*Ny

fig, axs = plt.subplots(nrows=3, ncols=5, figsize=(14, 9),
                        subplot_kw={'xticks': [], 'yticks': []})

fig.subplots_adjust(left=0.03, right=0.97, hspace=0.3, wspace=0.05)

axs = axs.reshape(15, -1).squeeze()
print axs.shape

for i in range(0, data.shape[1]):
    dx_interp = data[:feature_size, i]
    dy_interp = data[feature_size:, i]

    dx_resized = dx_interp.reshape(Nx, Ny) 
    dy_resized = dy_interp.reshape(Nx, Ny)
    opt_image = vec_color_encoding(dx_resized, dy_resized, encoding='rgb')
    
    axs[i].imshow(opt_image)
    axs[i].set_title('T = ' + str(i))

plt.style.use('seaborn')
```
